```python
import matplotlib.pyplot as plt
x = [1,2,6,18]
y = [3,10,12,20]
plt.plot(x, y, color='red', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='green', markersize=12)
plt.ylim(1,22)
plt.xlim(1,20)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Some cool customizations!')
plt.show()
```
